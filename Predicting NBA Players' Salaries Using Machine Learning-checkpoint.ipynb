**Feature Engineering & Data Preprocessing**

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('NBA - Dataset.csv')
df
```

```python
df.info()
```

```python
df['b_day'] =pd.to_datetime(df['b_day'])
df
```

```python
import datetime as dt                                    #Converting Birthday into Age
now_year =dt.datetime.today().year #current year
df['age']=now_year-df.b_day.dt.year
df
```

```python
df['salary'] = df['salary'].str.replace(',', '')    #converting salary column into an integer number by removing '$' & ',' 
df['salary'] = df['salary'].str.replace('$', '') 
df['salary'] = df['salary'].astype(int) 
df
```

```python
df.corr()['salary']            #to what extend my dependent variable is correlated with other features
```

```python
sns.scatterplot(x='rating',y='salary',data=df)
```

```python
sns.scatterplot(x='age',y='salary',data=df)
```

```python
df.isnull().sum()
```

```python
df['team'] = df['team'].fillna('No team')                 #Dealing with missing data
df['college'] = df['college'].fillna('No college')
df
```

```python
df["jersey"] = df["jersey"].str[1:].astype("int8")          #adjusting jersey column by removing the '#'
df
```

```python
df['height'] = df['height'].str[-4:]                 #adjusting height column by extracting the height (float) number
df
```

```python
df['weight'] = df['weight'].str[-9:]                 #adjusting weight column by extracting the weight (float) number
df
```

```python
df['weight'] = df['weight'].str.replace('kg.', '')    #replacing 'kg.' with " "
df
```

```python
df['draft_peak'] = df['draft_peak'].str.replace('Undrafted', '0')  #replacing 'Undrafted' with "0"
df
```

```python
df['draft_round'] = df['draft_round'].replace(r'Undrafted', 0, regex=True)  #replacing 'Undrafted' with "0"
df
```

**Linear Regression Model**

```python
X = df.iloc[:, 1].values.reshape(-1,1)              #Reshaping data using array.reshape(-1, 1)
y = df.iloc[:, 8].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
#Testing only 30% out of dataset & training the rest which is 70% 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model.fit(X_train,y_train)              #Fitting the training data
```

```python
prediction = model.predict(X_test)     
prediction
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error #Evaluating the model 
```

```python
df['salary'].mean()                         #Original salary average 
```

```python
mean_absolute_error(y_test,prediction)     #MAE Score equals 4382730/8530080 = 0.5
```

```python
np.sqrt(mean_squared_error(y_test,prediction))   #RMSE Score equals 6276641/8530080 = 0.6
```

**Cross Validation**

```python
X = df.iloc[:, 1].values.reshape(-1,1)              #Reshaping data using array.reshape(-1, 1)
y = df.iloc[:, 8].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
#Testing only 30% out of dataset & training the rest which is 70% 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train,y_train)         #Fitting the training data
```

```python
X_train = scaler.transform(X_train)                     #Transforming X_train & X_test 

X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model_2 = Ridge(alpha=100)
```

```python
from sklearn.model_selection import cross_validate
```

```python
scores =cross_validate(model_2,X_train,y_train,scoring=['neg_mean_absolute_error','neg_mean_squared_error'],cv=10) #for 10 rows 
scores
```

```python
scores = pd.DataFrame(scores)
scores
```

```python
scores.mean()           #Taking a gander at the average of all previous columns 
```

```python
model_2 = Ridge(alpha=10)                #let's try to improve the model by setting alpha to 10
```

```python
scores =cross_validate(model_2,X_train,y_train,scoring=['neg_mean_absolute_error','neg_mean_squared_error'],cv=10)
scores
```

```python
scores = pd.DataFrame(scores)
scores
```

```python
scores.mean()           #Taking a gander at the average of all previous columns 
```

```python
model_2.fit(X_train,y_train)
```

```python
prediction_2 = model_2.predict(X_test)
prediction_2
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test,prediction_2))  #RMSE Score equals 6319439/8530080 = 0.7
```
